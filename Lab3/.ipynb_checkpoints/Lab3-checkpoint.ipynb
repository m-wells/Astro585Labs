### 2a

```julia
# generate a vector (1-d Array) of  1024 Float64 numbers
x=rand(1024);
```

```julia
# print this vector to STDOUT and time how long it takes
stdout_time=@elapsed(print(STDOUT,x))
```

```julia
print("Print time to STDOUT: ",stdout_time," seconds")
```

### 2b

```julia
# load ASCII Read/Write routines
include("ioascii.jl");
```

### 2c

```julia
filename="test.txt";

# write x to filename and time how long it takes
writetime=@elapsed(asciiWrite(filename,x))
print("Write time to file: ",writetime," seconds.\n")

# compare the writetime to the stdout_time
if writetime < stdout_time
    print("This is ",stdout_time/writetime," times faster than STDOUT.")
else
    print("This is ",writetime/stdout_time," times slower than STDOUT.")
end
```

```julia
# initialize an empty array to hold readtimes
readtime=zeros(Float64,10)
for i=1:10
    readtime[i]=@elapsed(asciiRead(filename))
end
using PyPlot;
plot(readtime,color"red");
xlabel("Run Number");
ylabel("Time to read in seconds");
```
