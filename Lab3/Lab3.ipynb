```julia
# initialize an empty array to hold writetimes
writetime=zeros(Float64,10);
for i=1:10
    # print this vector to STDOUT and time how long it takes
    writetime[i]=@elapsed(print(STDOUT,x));
end

# take avg of writetimes
stdout_time=mean(writetime)
```

### 2c

```julia
aV_file="ascii_vector_test.txt";

# initialize an empty array to hold writetimes
writetime=zeros(Float64,10);
for i=1:10
    # write x to file and time how long it takes
    writetime[i]=@elapsed(asciiWrite(aV_file,x));
end

# take avg of writetimes
writetime_asciiVector=mean(writetime)
print("Write time to file: ",writetime_asciiVector," seconds.\n")

# compare the writetime to the stdout_time
if writetime_asciiVector < stdout_time
    print("This is ",stdout_time/writetime_asciiVector," times faster than STDOUT.")
else
    print("This is ",writetime_asciiVector/stdout_time," times slower than STDOUT.")
end
```

```julia
# initialize an empty array to hold readtimes
readtime=zeros(Float64,10);
runnumber=linspace(1,10,10);
for i=1:10
    readtime[i]=@elapsed(asciiRead(aV_file));
end
readtime_asciiVector=mean(readtime[2:10]);
```

```julia
using PyPlot;
plot(runnumber,readtime,color="red");
xlabel("Run Number");
ylabel("Time to read in seconds");
```

The above plot shows that the first read from the file requires much more time than the subsequent runs. The time to read again is negligible compare to the first read.

### 2g

Each format has its benefits. ASCII is useful because it is human readable outside of the program. However, if speed and size are critical factors then binary is the best choice. HDF5 is a powerful format that allows for multiple variables to be writen to one file quite easily and in an organized fashion. HDF5 is roughly the same size as a binary file but the access times do suffer when compared to binary.
