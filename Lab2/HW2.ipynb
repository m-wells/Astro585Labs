```julia
function evals_per_sec(foo,N,runs)
    # foo: name of a function to time
    # N: number of N evaluations per run
    #     i.e. N=10^6 will cause funct to return how many millions of evals per second
    # runs: number of runs to use for average
    
    x = rand(N)
    avgtime=0.0
    
    for i=1:runs
        avgtime=avgtime+@elapsed foo(x)
    end
    
    avgtime=avgtime/runs
    
    println(foo,":    ",1./avgtime)
end
```

Create some functional calls for the four basic arithmetic operations

```julia
addfoo(a)=a.+a;
minusfoo(a)=a.-a;
timesfoo(a)=a.*a;
dividefoo(a)=a./a;
```

```julia
N=10^6;
runs=100;
#println(".+:    ",1./(@elapsed x.+x))
evals_per_sec(addfoo,N,runs)
evals_per_sec(minusfoo,N,runs)
evals_per_sec(timesfoo,N,runs)
evals_per_sec(dividefoo,N,runs)
evals_per_sec(sin,N,runs)
evals_per_sec(cos,N,runs)
evals_per_sec(log,N,runs)
```

```julia
srand(42);
Nobs = 100;
z = zeros(Nobs);
sigma = 2. * ones(Nobs);
y = z + sigma .* randn(Nobs);
normal_pdf(z, y, sigma) = exp(-0.5*((y-z)/sigma)^2)/(sqrt(2*pi)*sigma);

function log_likelihood(y::Array, sigma::Array, z::Array)
    n = length(y);
    sum = zero(y[1]);
    for i in 1:n
        sum = sum + log(normal_pdf(y[i],z[i],sigma[i]));
        end;
    return sum;
end
```

### 2a

```julia
function calc_time_log_likelihood(Nobs::Int,M::Int=1)
    srand(42);
    
    z = zeros(Nobs);
    sigma = 2. * ones(Nobs);
    y = z + sigma .* randn(Nobs);
    tic();
    for i in 1:M
        log_likelihood(y,sigma,z)
    end
    time=toq();
end
```

```julia
function log_likelihood_with_asserts(y::Array, sigma::Array, z::Array)
    @assert length(y)==length(z)
    @assert length(y)==length(sigma)
    n = length(y);
    sum = zero(y[1]);
    for i in 1:n
        sum = sum + log(normal_pdf(z[i],y[i],sigma[i]));
        end;
    return sum;
end
```

```julia
function calc_time_log_likelihood_with_asserts(Nobs::Int,M::Int=1)
    z = zeros(Nobs);
    sigma = 2. * ones(Nobs);
    y = z + sigma .* randn(Nobs);
    tic();
    for i in 1:M
        log_likelihood_with_asserts(y,sigma,z)
    end
    time=toq();
end
```

```julia
calc_time_log_likelihood_with_asserts(100,100)
```

The extra time it takes to do the asserts is inconsequential. Since the asserts are so useful (and cheap computationally) it would be wise to include them as often as needed. This will make it easier to stop and correct issues later.

```julia
type State
    time::Float64
    rx::Float64
    ry::Float64
    vx::Float64
    vy::Float64
end
```

```julia
dvx_dt(state)=-G*M*state.rx/((state.rx^2+state.ry^2)^1.5)
```

```julia
dvy_dt(state)=-G*M*state.ry/((state.rx^2+state.ry^2)^1.5)
```

```julia
function integrate_euler!(init_state::State,dt::Float64,duration::Float64)
    steps=iceil(duration/dt)
    statelist=[init_state]
    for i=2:steps+1
        nextstep=State(0,0,0,0,0)
        nextstep.time=statelist[i-1].time+dt
        r=sqrt(statelist[i-1].rx^2+statelist[i-1].ry^2)
        ax=-G*M*statelist[i-1].rx/(r^3)
        ay=-G*M*statelist[i-1].ry/(r^3)
        nextstep.vx=statelist[i-1].vx+dt*ax
        nextstep.vy=statelist[i-1].vy+dt*ay
        nextstep.rx=statelist[i-1].rx+dt*statelist[i-1].vx
        nextstep.ry=statelist[i-1].ry+dt*statelist[i-1].vy
        statelist=[statelist,nextstep]
    end
    return statelist
end
```

## 3b

```julia
G=1;
M=1;
time_elasped=3*2pi;# 3 orbits (with a GM of 1)
num_steps=600;# 200 steps per orbit
step_size=time_elasped/num_steps
initial_state=State(0,1,0,0,1);# state(time,rx,ry,vx,vy);
testparticle=integrate_euler!(initial_state,step_size,time_elasped);
```

```julia
function integrate_leapfrog!(init_state::State,dt::Float64,duration::Float64)
    steps=iceil(duration/dt)
    statelist=[init_state]
    for i=2:steps+1
        nextstep=State(0,0,0,0,0)#Create a new empty next step
        nextstep.time=statelist[i-1].time+dt#save the time to the step
        
        rxmid=statelist[i-1].rx+0.5*dt*statelist[i-1].vx
        rymid=statelist[i-1].ry+0.5*dt*statelist[i-1].vy
        rmid=sqrt(rxmid^2+rymid^2)
        
        axmid=-G*M*statelist[i-1].rx/(rmid^3)
        aymid=-G*M*statelist[i-1].ry/(rmid^3)
        
        nextstep.vx=statelist[i-1].vx+dt*axmid
        nextstep.vy=statelist[i-1].vy+dt*aymid
        nextstep.rx=rxmid+0.5*dt*nextstep.vx
        nextstep.ry=rymid+0.5*dt*nextstep.vy
        
        statelist=[statelist,nextstep]
    end
    return statelist
end
```
