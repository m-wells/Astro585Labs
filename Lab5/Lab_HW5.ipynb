```julia
matmult(atest,xtest)
```

## Part B

```julia
function matmultcol(A::Array{Float64,2},x::Array{Float64,1})
    xlen=length(x);
    result=zeros(xlen);
    for i=1:xlen
        for j=1:xlen
            result[i]=result[i]+A[i,j]*x[j];
        end
    end
    return result
end
```

```julia
matmultcol(atest,xtest)
```

## Part C

```julia
function matmultrow(A::Array{Float64,2},x::Array{Float64,1})
    xlen=length(x);
    result=zeros(xlen);
    for i=1:xlen
        for j=1:xlen
            result[j]=result[j]+A[j,i]*x[i];
        end
    end
    return result
end
```

```julia
matmultrow(atest,xtest)
```

## Part D

We believe Julia's built procedure will be the best one. Between part b and c the function in part c should perform faster because we are incrementing rows in the inner loops where are elements are closer together in memory. The function in part b has to jump farther in memory to access the elements for the inner loop.

## Part E

For small matrices the performance should be comparable because all the data elements are now closer together in memory.

## Part F

```julia
abig=rand(1024,1024)
asmall=rand(8,8)
xbig=rand(1024)
xsmall=rand(8)
println("julia's default on big matrix:  ", @elapsed x=matmult(abig,xbig))
println("matmultcol on big matrix:  ", @elapsed x=matmultcol(abig,xbig))
println("matmultrow on big matrix:  ", @elapsed x=matmultrow(abig,xbig))
println("julia's default on small matrix:  ", @elapsed x=matmult(asmall,xsmall))
println("matmultcol on small matrix:  ", @elapsed x=matmultcol(asmall,xsmall))
println("matmultrow on small matrix:  ", @elapsed x=matmultrow(asmall,xsmall))
```

## Part G

We find that the performance of the function of part c, where the inner loop runs over rows, is consistently better than the one in part b, which matches our predictions. Julia's default function has comparable performantce to part c.
